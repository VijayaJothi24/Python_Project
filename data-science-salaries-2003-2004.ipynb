```python
# Check for count
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'employee_residence': ['US', 'GB', 'CA', 'ES', 'IN', 'DE', 'FR', 'PT', 'BR', 'GR'],
   'company_location': ['US', 'GB', 'CA', 'ES', 'IN', 'DE', 'FR', 'BR', 'AU', 'GR']
})

# Function to find common words
def count_common_words(row):
    words1 = set(row[ 'employee_residence'].split())
    words2 = set(row[ 'company_location'].split())
    return len(words1 & words2)

# Apply the function to each row
df['CommonWordCount'] = df.apply(count_common_words, axis=1)

print(df)
```

```python
# Importing library and dataframe

import pandas as pd
df=pd.read_csv('/kaggle/input/data-science-salaries-2023/ds_salaries.csv')
```

```python
#importing seaborn library
import seaborn as sns
```

```python
 # Importing matplotlib library 
import matplotlib.pyplot as plt
```

```python
df
```

```python
job_counts=df['job_title'].value_counts()
# Find the top 10 most common job titles
top_10jobs=df['job_title'].value_counts().head(10).index
#Filter the dataset for these top 10 job titles
top_jobs_df=df[df['job_title'].isin(top_10jobs)]
average_salary_top_jobs= top_jobs_df.groupby('job_title')['salary_in_usd'].mean().round(2).reset_index()
average_salary_top_jobs
```

```python
sal_mean_top=average_salary_top_jobs.reset.index()
sal_mean_top_sorted=sal_mean_top.sort_values(by='salary_in_usd')
```

```python
sal_mean_top_sorted
```

```python
sns.set_style('dark')
plt.figure(figsize=(8,6))
sns.barplot(x='salary_in_usd',y='job_title',data=average_salary_top_jobs, palette='Blues_d',orient='h')
plt.title('Average Salary for the Top 10 Most Common Jobs',fontsize=16)
plt.xlabel('Job Title',fontsize=14)
plt.ylabel('Average Salary',fontsize=14)
```

```python
sns.set_style('dark')
plt.figure(figsize=(8,6))
sns.barplot(x='salary_in_usd',y='job_title',data=sal_mean_top_sorted, palette='Blues_d',orient='h')
plt.title('Average Salary for the Top 10 Most Common Jobs',fontsize=16)
plt.xlabel('Job Title',fontsize=14)
plt.ylabel('Average Salary',fontsize=14)
```
